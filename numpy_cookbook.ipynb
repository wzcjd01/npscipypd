# Advanced Indexing and Array Concepts

```python
%matplotlib inline
```

```python
import scipy.misc
import matplotlib.pyplot as plt
```

```python
lena = scipy.misc.lena()
acopy = lena.copy()
aview = lena.view()
```

```python
# Plot Lena arry
plt.subplot(221)
plt.imshow(lena)

# Plot Lena copy
plt.subplot(222)
plt.imshow(acopy)

# Plot Lena view
plt.subplot(223)
plt.imshow(aview)

# plot the view after changes
aview[:, :] = 0
plt.subplot(224)
plt.imshow(aview)
 
plt.show()
```
